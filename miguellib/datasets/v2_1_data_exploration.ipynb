# Dataset Exploration

For the final recommender we decided to upgrade the dataset to a more complete one. This is the data exploration of it.

Dataset Name: SkinSAFE Skincare Products (50,000+)

Kaggle Link: https://www.kaggle.com/datasets/kazireyazulhasan/skinsafe-skincare-db-50000-products?resource=download

```python
BASE_DIR = Path().resolve() 
DATASET_PATH = BASE_DIR / ".." / ".." / "MiguelLib" / "datasets" / "data" / "products_raw.csv"
DATASET_PATH = DATASET_PATH.resolve()


df = pd.read_csv(DATASET_PATH)
df.head()
```

### Overview

```python
print("=== Dataset Overview ===")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
print("\nColumns:")
for col in df.columns:
    print(f" - {col} ({df[col].dtype})")
```

```python
print("\n=== Missing Values ===")
missing_counts = df.isnull().sum()
missing_pct = (df.isnull().mean() * 100).round(2)
missing_df = pd.DataFrame({"count": missing_counts, "percent": missing_pct})
display(missing_df)

dupes = df.duplicated(subset=["product_name", "brand"]).sum()
print(f"\nDuplicate rows (product_name + brand): {dupes}")
```

```python
dupes = df.duplicated(subset=["product_name", "brand", "ingredients"]).sum()
print(f"\nDuplicate rows (product_name + brand + ingredients): {dupes}")
```

```python
duplicate_rows = df[df.duplicated(subset=["product_name", "brand", "ingredients"], keep=False)]
duplicate_rows = duplicate_rows.sort_values(["product_name", "brand"]).reset_index(drop=True)
display(duplicate_rows.head(20))  
```

## Why is this a good dataset to train our recommender with?

This dataset is highly suitable for our skincare recommender because it is comprehensive, covering 50,000+ products across hundreds of brands, categories, and usage types. It includes detailed ingredient lists in standard format, which allows us to match products to user preferences and skin concerns effectively. The dataset is clean and even though it has 600+ duplicates it is nothing that preprocessing can't handle. Most importantly, the ingredient lists are extensive and seem to follow the same format (INCI). This dataset provides the necessary resources to create a relevant skincare recommender that covers a range of products and preferences.
